```python
import os
import re
import datetime

def rename_files(folder_path):
    file_list = os.listdir(folder_path)
    
    for file_name in file_list:
        old_file_path = os.path.join(folder_path, file_name)
        
        # Extract the date and time from the file name
        match = re.search(r'\d{2}:\d{2}, (\d{2}-\d{2}-\d{4}|\d{2} \w+ \d{4})', file_name)
        
        if match:
            date_string = match.group(1)
            
            # Parse the date string into a datetime object
            if "-" in date_string:
                date = datetime.datetime.strptime(date_string, "%d-%m-%Y")
            else:
                date = datetime.datetime.strptime(date_string, "%d %B %Y")
            
            # Format the date as DD-MM-YYYY
            new_date_string = date.strftime("%d-%m-%Y")
            
            # Create the new file name
            new_file_name = file_name.replace(date_string, new_date_string)
            new_file_path = os.path.join(folder_path, new_file_name)
            
            # Rename the file
            os.rename(old_file_path, new_file_path)
            print(f"Renamed {file_name} to {new_file_name}")

# Usage example
folder_path = "/home/jc/Documents/GitHub/ELLM/ELLM-ALRIER/Datas"
rename_files(folder_path)
```
